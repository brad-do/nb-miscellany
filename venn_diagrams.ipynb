### Who doesn't have a need to create a nice [Venn Diagram](https://en.wikipedia.org/wiki/Venn_diagram) from time to time?  Recently, one of my kids had a homework assignment to create a Venn Diagram showing the three branches of the federal government and providing examples where the branches overlap.  I came up with the following to help her:

### Step 1: Install matplotlib-venn
There's an excellent extension to matplotlib called [matplotlib-venn](https://pypi.org/project/matplotlib-venn/).  I installed this package to my environment with conda:
```
conda install -c conda-forge matplotlib-venn
```

### Step 2: Load the necessary packages

````python
import matplotlib.pyplot as plt
from matplotlib_venn import venn3
import matplotlib.patheffects as path_effects

%matplotlib inline
````

### Step 3: Generate your Venn Diagrams
I've had to play around with the package a little and do some googling, but it's slowly starting to make some sense ([this site](https://python-graph-gallery.com/venn-diagram/) was especially helpful).  Here's a generic example to demonstrate how you can shape your diagram and apply some labels.  By default, the number values of each section of the circles will appear.

````python
# Make the diagram
fig, ax = plt.subplots(figsize=(8, 8))
v = venn3(subsets = (10, 10, 4, 10, 4, 4, 2), set_labels = ('Set1', 'Set2', 'Set3'), ax=ax)
v.get_label_by_id('100').set_text('Something1')
v.get_label_by_id('010').set_text('Something2')
v.get_label_by_id('001').set_text('Something3')
v.get_label_by_id('110').set_text('Some1\nSome2')
v.get_label_by_id('011').set_text('Some2\nSome3')
v.get_label_by_id('101').set_text('Some1\nSome3')
v.get_label_by_id('111').set_text('Unicorn')
plt.title("Generic Venn diagram example")

plt.show()
````

````python
fig, ax = plt.subplots(figsize=(10, 10), sharey=True)
v = venn3(subsets = (10, 10, 4, 10, 4, 4, 2), set_labels = ('', '', ''), ax=ax)
v.get_label_by_id('100').set_text('Executive')
v.get_label_by_id('010').set_text('Legislative')
v.get_label_by_id('001').set_text('Judicial')
v.get_label_by_id('110').set_text('Example 1')
v.get_label_by_id('011').set_text('Example 2')
v.get_label_by_id('101').set_text('Example 3')
v.get_label_by_id('111').set_text('')
plt.title("The Three Branches of the US Government")

example_text = ('Example 1: The Vice President is considered "President of the Senate" and can vote to break ties.\n'
                'Example 2: The Legislature confirms Supreme Court justices.\n'
                'Example 3: The Executive appoints potential Supreme Court justices.')

text = fig.text(0.0, 0.05, example_text, ha='left', va='bottom', size=14)
text.set_path_effects([path_effects.Normal()])

plt.show()
````
