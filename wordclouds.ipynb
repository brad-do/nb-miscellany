```python
import pandas as pd
import json
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
with open('./data/activity-stream.discovery_stream.json', 'rb') as f:
    data = json.load(f)
    
domains = []
for r in data['feeds']['https://getpocket.cdn.mozilla.net/v3/firefox/global-recs?version=3&consumer_key=$apiKey&locale_lang=$locale&count=30']['data']['recommendations']:
    domains.append([r['domain']])
    
df_domains = pd.DataFrame(domains, columns=['domain'])
df_domains.head()
```

```python
text = ' '.join(df_domains.domain.tolist())

wordcloud = WordCloud().generate(text)
_ = plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
_ = plt.title('Random Domains')
```

```python
wordcloud.words_
```

```python
df_domains.domain.value_counts().to_dict()
```

```python
freqs = df_domains.domain.value_counts().to_dict()

wordcloud = WordCloud().generate_from_frequencies(freqs)
_ = plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
_ = plt.title('Random Domains')
```
